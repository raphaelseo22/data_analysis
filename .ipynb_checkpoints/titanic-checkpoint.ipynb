```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic_train.csv")
```

```python
df
```

```python
df.Sex.value_counts()
```

```python
sex_num = {"Sex" : {"male":0, "female":1}}
```

```python
df1 = df.replace(sex_num)
```

```python
df1
```

```python
df1.groupby("Survived").mean()
```

### Age 결측값을 maen으로 채웠을 경우 correlation

```python
df2 = df1.copy()
```

```python
df2.Age = df2.Age.fillna(round(df1.Age.mean()))
```

```python
df2.Age
```

```python
df2.corr()
```

### Age 결측값을 median으로 채웠을 경우 correlation

```python
df3 = df1.copy()
```

```python
df3.Age = df3.Age.fillna(df1.Age.median())
```

```python
df3.corr()
```

```python
df.dtypes
```

```python
df_num = df1.select_dtypes(include=["int64", "float64"])
```

```python
df_num
```

```python
for feature_nm in df_num.columns : 
    if not feature_nm == "Survived":
        plt.title(feature_nm)
        for cls in df_num.Survived.unique() : 
            vec = df_num.loc[df.Survived == cls, feature_nm].values
            plt.hist(vec, alpha=.3, label=cls)
        plt.legend()
        plt.grid()
        plt.show()
```

- 여성일 경우 남성보다 생존할 확률이 높음

```python
plt.title("")
plt.hist(df_num.loc[(df_num.Sex == 0), "Age"].values, alpha= 0.3, label="all")
plt.hist(df_num.loc[(df_num.Sex == 0) & (df_num.Survived == 1), "Age"].values, alpha= 0.3, label="survive")
plt.legend()
plt.grid()
plt.show()
```

```python
len(df_num.loc[(df_num.Sex == 0) & (df_num.Survived == 1), "Age"].values)/len(df_num.loc[(df_num.Sex == 0), "Age"].values) * 100
```

- 남성의 경우 데이터의 약 18.89% 가 생존

```python
plt.title("")
plt.hist(df_num.loc[(df_num.Sex == 1), "Age"].values, alpha= 0.3, label="all")
plt.hist(df_num.loc[(df_num.Sex == 1) & (df_num.Survived == 1), "Age"].values, alpha= 0.3, label="survive")
plt.legend()
plt.grid()
plt.show()
```

```python
len(df_num.loc[(df_num.Sex == 1) & (df_num.Survived == 1), "Age"].values)/len(df_num.loc[(df_num.Sex == 1), "Age"].values) * 100
```

- 여성의 경우 데이터의 약 74.2% 가 생존
