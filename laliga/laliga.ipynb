```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("lalaiga_all_player.csv")
```

```python
# null 값은 관측되지 않은 값이므로 0으로 대체
data = data.fillna(0)
```

```python
data
```

```python
# 출전수를 주전과 교체로 분리
starter = []
sub = []
for idx, app in enumerate(data["App"]):
    if type(app) == int:
        print(app)
        print(idx)
    else:
        if "(" in app:
            start = app.index("(")
            end = app.index(")")
            starter.append(app[:start])
            sub.append(app[start:end])
        else:
            starter.append(app)
            sub.append(0)
```
